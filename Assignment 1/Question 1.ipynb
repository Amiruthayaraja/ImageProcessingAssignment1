```python
import numpy as np
import cv2 as cv
import math 
from matplotlib import pyplot as plt
```

```python
image = cv.imread('bird.jpg')
levels = int(input("Number of desired Intensity level: "))
div = 255/(levels - 1)
image1 = np.round(image/div)*div
plt.subplot(121),plt.imshow(image),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(image1),plt.title('Reduced Intensity level Image')
plt.xticks([]), plt.yticks([])
cv.imwrite('Image1.jpg', image1)
```
