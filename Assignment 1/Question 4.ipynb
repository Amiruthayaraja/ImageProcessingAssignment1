```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('bird.jpg')
kernel3 = np.ones((3,3),np.float32)/9
kernel5 = np.ones((5,5),np.float32)/25
kernel7 = np.ones((7,7),np.float32)/49
dst3 = cv.filter2D(img,-1,kernel3)
dst5 = cv.filter2D(img,-1,kernel5)
dst7 = cv.filter2D(img,-1,kernel7)
plt.subplot(141),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(142),plt.imshow(dst3),plt.title('Averaging 3')
plt.xticks([]), plt.yticks([])
plt.subplot(143),plt.imshow(dst5),plt.title('Averaging 5')
plt.xticks([]), plt.yticks([])
plt.subplot(144),plt.imshow(dst7),plt.title('Averaging 7')
plt.xticks([]), plt.yticks([])
plt.show()

cv.imwrite('Image4_3x3.jpg', dst3)
cv.imwrite('Image4_5x5.jpg', dst5)
cv.imwrite('Image4_7x7.jpg', dst7)
```
