```python
import numpy as np
import cv2 as cv
import math 
from matplotlib import pyplot as plt
```

```python
img = cv.imread('bird.jpg')
kernel3 = np.ones((3,3),np.float32)/9
kernel10 = np.ones((10,10),np.float32)/100
kernel20 = np.ones((20,20),np.float32)/400
dst3 = cv.filter2D(img,-1,kernel3)
dst10 = cv.filter2D(img,-1,kernel10)
dst20 = cv.filter2D(img,-1,kernel20)
plt.subplot(141),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(142),plt.imshow(dst3),plt.title('3x3 Kernel Image')
plt.xticks([]), plt.yticks([])
plt.subplot(143),plt.imshow(dst10),plt.title('10x10 Kernel Image')
plt.xticks([]), plt.yticks([])
plt.subplot(144),plt.imshow(dst20),plt.title('20x20 Kernel Image')
plt.xticks([]), plt.yticks([])
plt.show()

cv.imwrite('Image2_3x3.jpg', dst3)
cv.imwrite('Image2_10x10.jpg', dst10)
cv.imwrite('Image2_20x220.jpg', dst20)
```
