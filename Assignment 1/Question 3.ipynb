```python
import cv2
from matplotlib import pyplot as plt
image = cv2.imread('bird.jpg')
h, w, c = image.shape
image_90_rotated = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
M = cv2.getRotationMatrix2D((h/2,w/2),-45,1) 
image_45_rotated = cv2.warpAffine(image,M,(h,w))
plt.subplot(121),plt.imshow(image_90_rotated),plt.title('Image Rotated 90')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(image_45_rotated),plt.title('Image Rotated 45')
plt.xticks([]), plt.yticks([])
cv2.imwrite('rotated90.jpg', image_90_rotated)
cv2.imwrite('rotated45.jpg', image_45_rotated)
```
